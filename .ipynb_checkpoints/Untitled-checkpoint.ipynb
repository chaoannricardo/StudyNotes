```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
womenMenMeans = (45, 67, 64, 55, 52)
menTest = (20, 35, 30, 35, 27)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
menTestStd = (4, 6, 2, 3, 1)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence
```

```python
fig, ax = plt.subplots()

# p1 = ax.bar(ind, menMeans, width, yerr=menStd, label='Men')
# p2 = ax.bar(ind, womenMeans, width,
#             bottom=menMeans, yerr=womenStd, label='Women')
# p3 = ax.bar(ind, menTest, width,
#             bottom=womenMeans, yerr=menTestStd, label='MenTest')

p1 = ax.bar(ind, menMeans, width, label='Men')
p2 = ax.bar(ind, womenMeans, width, bottom=menMeans, label='Women')
p3 = ax.bar(ind, menTest, width, bottom=womenMenMeans, label='MenTest')

ax.axhline(0, color='grey', linewidth=0.8)
# ax.set_ylabel('Scores')
# ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend()

# Label with label_type 'center' instead of the default 'edge'
ax.bar_label(p1, label_type='edge')
ax.bar_label(p2, label_type='edge')
ax.bar_label(p3, label_type='edge')
ax.bar_label(p3)

plt.show()
```
