## Import Data with applicable columns

```python
# import ok csv data, only with applicable columns
import pandas as pd

df = pd.read_csv("C:/Users/1907075/Project/VAS_Project_Python/all_0720_NG.csv")
newDF = pd.DataFrame({'DateTime24': df.loc[:,'DateTime24'],
                      'FileName': df.loc[:,'filename'],
                     'VG001': df.loc[:,'VG001'],
                      'VG002': df.loc[:,'VG002'],
                     'VG003': df.loc[:,'VG003'],
                     'VG111': df.loc[:,'VG111'],
                     'VG921': df.loc[:,'VG921'],
                     'VG922': df.loc[:,'VG922'],
                     'VG923': df.loc[:,'VG923']})

newDF.loc[:,'Result'] = 'NG'
```

```python
# Show the head of the data
print(newDF.head())
print(newDF.describe())
```

## Draw scatter plot of each column

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(100, 100))
subplot_iterator = 1
column_list = ['VG001','VG002','VG003','VG111','VG921','VG922','VG923']

for i,j in enumerate(column_list):
    for a,b in enumerate(column_list):
        if i != a:
            fig.add_subplot(16,4, subplot_iterator)
            Xlist = newDF.loc[:,j].values.tolist()
            Ylist = newDF.loc[:,b].values.tolist()
            label = str(j) + ' X ' + str(b)
            plt.scatter(Xlist, Ylist, color='blue', label = label, s=1)
            plt.legend()
            plt.xlabel(str(j))
            plt.ylabel(str(b))
            subplot_iterator = subplot_iterator + 1

#plt.show()
plt.savefig('Hello123.png')
```

## Divide data with distinct filename (RNN Preprocessing)

```python
import pandas as pd 

filename_list_unique = newDF.loc[:,'FileName'].unique()  # distinct values within 'FileName' column.

for i,j in enumerate(filename_list_unique):
    saveDF = pd.DataFrame() # intitate a new dataframe to store saved data.
    saveFileName = 'C:/Users/1907075/Project/VAS_Project_Python/RNN_data/RNN_Data_' + str(j) # set save file name as 'RNN Data' + FileName.
    for a,b in enumerate(newDF.loc[:,'FileName']):
        if j == b: # if the row value matches the value on distinct list.
            saveDF = saveDF.append(newDF.iloc[a,:]) # append the row into the dataframe.
    saveDF.to_csv(saveFileName, sep=',', encoding='utf_8', index=False)
```
