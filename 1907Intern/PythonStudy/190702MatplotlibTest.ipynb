### Plot Test

```python
import matplotlib.pyplot as plt

# Data1
listx1 = [1,5,7,9,13,16]
listy1 = [15,50,80,40,70,50]

# Plot1
plt.plot(listx1, listy1, label="Male")

# Data2
listx2 = [2,6,8,11,14,16]
listy2 = [10,40,30,50,80,60]

# Plot2
plt.plot(listx2, listy2, color='red', linewidth=5, linestyle='--', label = 'Female')

# legend
plt.legend()

#Limits
plt.xlim(0,20)
plt.ylim(0,100)

plt.title("Pocket money")
plt.xlabel("Age")
plt.ylabel("Money")
```

### Bar Plot Test

```python
plt.bar(listx1, listy1, label='Male')
plt.bar(listx2, listy2, label='Female')
plt.title("Pocket Amount of each sex")
plt.xlabel("age")
plt.ylabel("pocket money")
```

### Pie Plot Test

```python
import matplotlib.pyplot as plt

labels = ["East Region", "South Region", "North Region", "Central Region"]
sizes = [5, 10, 20, 15]
colors = ['red', 'green', 'blue', 'orange']
explode = (0, 0, 0.05, 0)
plt.pie(sizes, explode = explode, labels = labels, colors = colors,
       labeldistance = 1.1, autopct = '%3.1f%%', shadow = 'True', 
       startangle = 90, pctdistance = 0.6)
plt.axis('equal')
plt.legend()
plt.show()
```
