```python
y=pd.read_csv("C:/Users/Student/Desktop/crazy/Y.csv")
Y=y.iloc[:,1:4]
User=pd.read_csv("C:/Users/Student/Desktop/crazy/user.csv")
user=User.iloc[:,1:-1]
Q=Y.iloc[:,1:4]
```

```python
business_emotion_result=pd.read_csv("C:/Users/Student/Desktop/crazy/business_emotion_result.csv")
business_emotion_result=business_emotion_result.iloc[:,1:9]
business_timestamp_flatten=pd.read_csv("C:/Users/Student/Desktop/crazy/business_timestamp_flatten.csv")
business_timestamp_flatten=business_timestamp_flatten.iloc[:,1:9]
business_category_flatten=pd.read_csv("C:/Users/Student/Desktop/crazy/business_category_flatten.csv")
business_category_flatten=business_category_flatten.iloc[:,1:99]
business_review_count=pd.read_csv("C:/Users/Student/Desktop/crazy/business_review_count.csv")
business_review_count=business_review_count.iloc[:,1:4]
business_received_average_star=pd.read_csv("C:/Users/Student/Desktop/crazy/business_given_average_star.csv")
business_received_average_star=business_received_average_star.iloc[:,1:3]
```

```python
business_timestamp_flatten
```

```python
x=pd.merge(Y,business_emotion_result,on=["review_id"],how='right')
f=pd.merge(x,user,on=["review_id"],how='right')
```

```python
r=pd.merge(f,key,on=["review_id"],how='left')
r.isnull().values.any()
```

```python
t=pd.merge(r,business_received_average_star,on=["business_id"],how='left')
t.isnull().values.any()
t
```

```python
u=pd.merge(t,business_review_count,on=["business_id"],how='left')
u.isnull().values.any()
u.shape

```

```python
p=pd.merge(u,business_timestamp_flatten,on=["business_id"],how='left')
p=p.dropna(axis=0)
p.isnull().values.any()
```

```python
FINAL=pd.merge(p,business_category_flatten,on=["business_id"],how='left')
FINAL.isnull().values.any()
FINAL.shape
```
